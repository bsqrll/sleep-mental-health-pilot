# Sono e Saúde Mental

**Pergunta:**
Dormir mal está associado a pior bem-estar psicológico?

**Estratégia:**
Usar horas de sono e percepção de sono suficiente como variáveis explicativas
e cansaço diário como proxy de bem-estar psicológico.

**Observação:**
Análise de associação (não causal).


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid", palette="colorblind")
```

```python
df = pd.read_csv("../data/SleepStudyPilot.csv")
df.head()
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.describe()
```

```python
df["Enough"].value_counts()
```

## Percepção de sono e cansaço diário

```python
sns.boxplot(x="Enough", y="Tired", data=df)
plt.title("Cansaço diário vs percepção de sono suficiente")
plt.xlabel("Acha que dorme o suficiente?")
plt.ylabel("Cansaço diário")
plt.show()
```

### Percepção de sono e cansaço diário

Embora a mediana de cansaço seja semelhante entre os grupos,
os participantes que relatam não dormir o suficiente apresentam
maior variabilidade no cansaço diário, incluindo níveis mais elevados
de exaustão. Esse padrão sugere que a percepção subjetiva do sono
pode estar relacionada à experiência de bem-estar, para além
da quantidade de horas dormidas.

## Quantidade de sono e cansaço

```python
sns.pointplot(x="Hours", y="Tired", data=df)
plt.title("Cansaço médio por horas de sono")
plt.show()
```

A relação entre horas de sono e cansaço diário não apresenta
um padrão visual claro no estudo piloto. Mesmo entre participantes
que dormem a mesma quantidade de horas, observa-se grande variação
nos níveis de cansaço relatados. Esse resultado sugere que a
quantidade de sono, isoladamente, pode não ser suficiente para
explicar o bem-estar percebido.

## Uso do celular antes de dormir

```python
df["PhoneTime"].value_counts()
```

No estudo piloto, todos os participantes relataram utilizar o celular
antes de dormir. Dessa forma, essa variável não apresenta variabilidade
suficiente para permitir comparações com o cansaço diário.
Esse resultado sugere que o uso do celular antes de dormir pode ser
um comportamento amplamente disseminado no grupo analisado,
dificultando sua avaliação como fator diferenciador do bem-estar.
